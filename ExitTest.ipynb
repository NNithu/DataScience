# outliers Detection

```python
# Select numerical columns with potential outliers
numerical_columns = ['age'
                     ,'ball_controlling_skills', 'body_reflexes', 'body_balance',
                     'jumping_skills', 'penalties_conversion_rate', 'mental_strength', 'goalkeeping_skills',
                     'defending_skills', 'passing_skills', 'dribbling_skills', 'body_strength_stamina',
                     'max_running_speed', 'behaviour_rating', 'matches_played', 'fitness_rating',
                      'no_of_disqualifications']
```

```python
plt.figure(figsize=(20, 15))
for column in numerical_columns:
    plt.subplot(4, 6, numerical_columns.index(column) + 1)
    sns.boxplot(data[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
from scipy.stats import zscore

# Calculate z-scores for numerical columns
z_scores = data[numerical_columns].apply(zscore)

# Identify data points with z-scores above a threshold (e.g., 3)
threshold = 3
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Print rows containing outliers
outliers_rows = data[outliers.any(axis=1)]
print(outliers_rows)
```

# Handling outliers


```python
# Calculate IQR for numerical columns
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for identifying outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap outliers in each numerical column
for col in numerical_columns:
    data[col] = np.where(data[col] < lower_bound[col], lower_bound[col], data[col])
    data[col] = np.where(data[col] > upper_bound[col], upper_bound[col], data[col])
```

```python
plt.figure(figsize=(20, 15))
for column in numerical_columns:
    plt.subplot(4, 6, numerical_columns.index(column) + 1)
    sns.boxplot(data[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
# Data preprocessing: Convert height from feet' inches" format to centimeters
def feet_inches_to_cm(height):
    feet, inches = map(int, height.split("'"))
    total_inches = feet * 12 + inches
    cm = total_inches * 2.54
    return cm
```

```python

data['height_cm'] = data['height'].apply(feet_inches_to_cm)
```

```python
# Drop the original 'height' column
data.drop('height', axis=1, inplace=True)
```

# Feature Engineering:

```python
#creating age group from age
def categorize_age(age):
    if age < 18:
        return 'Teen'
    elif 18 <= age < 30:
        return 'Young Adult'
    elif 30 <= age < 50:
        return 'Adult'
    else:
        return 'Senior'

data['age_group'] = data['age'].apply(categorize_age)
```

```python
# Drop the original 'age' column
data.drop('age', axis=1, inplace=True)
```

```python
#BMI (Body Mass Index): calucuation from weight and height
data['bmi'] = data['weight'] / ((data['height_cm'] / 100) ** 2)
```

```python
# Drop the original 'height_cm' column
data.drop('height_cm', axis=1, inplace=True)
```

# Handling Categorical Data: Convert categorical variables using one-hot encoding.

```python
categorical_columns = ['gender', 'strong_foot', 'dedication_level', 'coaching', 'age_group']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
#Scaling
import pandas as pd
from sklearn.preprocessing import StandardScaler



# Separate the target variable ('selection') from the features
# X = data.drop('selection', axis=1)
# Select features and target
X = data.drop(columns=['selection', 'player_id', 'name','country'])  # Drop player_id and name
y = data['selection']

# Get the column names of the numeric features
numeric_features = X.select_dtypes(include='number').columns

# Create the StandardScaler object
scaler = StandardScaler()

# Standardize the numeric features
X[numeric_features] = scaler.fit_transform(X[numeric_features])

# Now the numeric features in X are standardized with mean=0 and std=1
```

# # Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Modeling

```python
#import models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

```

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC()
}
```

```python

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{model_name} Accuracy: {accuracy:.2f}')
```

most accurancy is for random forest model so we selcet that model

#  hyperparameter tuning for a Random Forest
 # Fine tunning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define the parameter grid to search through
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2)

# Fit the grid search to your training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Get the best model
best_rf_model = grid_search.best_estimator_

# Evaluate the model on the test set
y_pred = best_rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on test set:", accuracy)
```
