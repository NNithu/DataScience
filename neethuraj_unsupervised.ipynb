```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

# 1. K-means Clustering:

```python
# Read the CSV file
data = pd.read_csv("/home/neethu/Downloads/Wine_clust .csv")
```

```python
# Preprocess the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# Find the optimal number of clusters (K)
distortions = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(scaled_data)
    distortions.append(kmeans.inertia_)
```

```python
# Plot the elbow curve to determine the optimal K
import matplotlib.pyplot as plt
plt.plot(range(1, 10), distortions, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Distortion')
plt.title('K-means Elbow Curve')
plt.show()
```

```python
# Choose the optimal K and create the K-means model
optimal_k = 3  #  K  determined from graph knee point wherev curve stats to flatten
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
kmeans.fit(scaled_data)
```

```python
# Access the cluster labels and centroids
cluster_labels = kmeans.labels_
```

```python
cluster_labels
```

```python
centroids = kmeans.cluster_centers_
```

# 2. Hierarchical Clustering:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Read the CSV file
data1 = pd.read_csv("/home/neethu/Downloads/Wine_clust .csv")
```

```python
# Preprocess the data
scaler = StandardScaler()
scaled_data1 = scaler.fit_transform(data1)
```

```python
# Perform hierarchical clustering
linkage_matrix = linkage(scaled_data1, method='ward')
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
# Choose the optimal number of clusters and create the hierarchical clustering model
optimal_clusters1 = 3  #   optimal number of clusters determined
from sklearn.cluster import AgglomerativeClustering
hierarchical = AgglomerativeClustering(n_clusters=optimal_clusters1, affinity='euclidean', linkage='ward')
cluster_labels1 = hierarchical.fit_predict(scaled_data1)
```

```python
cluster_labels1
```

```python
# Calculate the centroids
centroids1 = []
num_clusters=3
for cluster in range(num_clusters):
    cluster_data1 = data[cluster_labels1 == cluster]
    cluster_centroid1 = np.mean(cluster_data1, axis=0)  # or np.median(cluster_data, axis=0)
    centroids1.append(cluster_centroid1)
centroids1 = np.array(centroids1)
```

#  DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Read the CSV file
data2 = pd.read_csv("/home/neethu/Downloads/Wine_clust .csv")
```

```python
# Preprocess the data
scaler = StandardScaler()
scaled_data2 = scaler.fit_transform(data2)
```

```python
# Create and fit the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)
cluster_labels2 = dbscan.fit_predict(scaled_data2)
```

```python
# Access the cluster labels (-1 represents outliers/noise)
unique_labels = np.unique(cluster_labels2)
```

```python
unique_labels
```

```python
# Print the number of clusters and outliers
n_clusters = len(unique_labels) - 1
n_noise = np.sum(cluster_labels2 == -1)
print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```
